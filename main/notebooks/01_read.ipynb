```python
values_null = df[df.isnull().any(axis=1)]
print(values_null)
```

```python
df = df.dropna(subset=['Duration (min)', 'Review', 'Cast', 'Review Title', 'Year', 'Votes'])
df.info()
```

```python
df['metascore_missing'] = df['Metascore'].isna().astype(int)    # comprueba fila a fila si el dato metascore es nan, si falta -> true, en caso contrario, false, .astype devuelve la lista convirtiendo los valores booleanos a números enteros (0/1)
df['Metascore'].fillna(df['Metascore'].median(), inplace=True)  # calcula la mediana para el valor metascore faltante, es decir, NaN, con inplace=True, le damos permiso para modificarlo directamente

df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df.head()
df.info()
```
